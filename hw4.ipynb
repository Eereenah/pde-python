## Problem 1

\begin{cases}
u_t = 4u_{xx}, \ x\in (0, 1), \ t>0 \\[6pt]
u(x, 0) = x^2(1-x), \ x \in [0, 1] &(i)\\[6pt] 
u(0, t) = 0, \ t>0  &(ii)\\[6pt] 
u(1, t) = 0, \ t>0  &(iii)\\[6pt]
\end{cases}

### Plot 1D graph of $u(x, 0), u(x, 1), u(x, 10)$ for $x \in [0, 1]$ on the same graph

The solution of the IBVP will be $u(x,t) = \sum_{m=1}^\infty(\frac{8(-1)^{m+1}-4}{\pi^3m^3})\sin\left(\pi mx\right)e^{-4m^2\pi^2 t}$. The plots will be the following:

```python
def f1(x, t):
    n = 10
    s = [(8*(-1)**(k+1)-4)/(np.pi**3)*np.sin(np.pi*k*x)*np.exp(-4*k**2*np.pi**2*t) for k in range(1, n+1)]
    return sum(s)
```

```python
z1 = np.linspace(0, 1, 50)
plt.plot(z1, f1(z1, 0), z1, f1(z1, 1), z1, f1(z1, 10))
plt.show()
```

The solution of the IBVP will be $u(x,t) = -\frac{x}{2} + 1 +(\sum_{m=1}^\infty(\frac{-2}{\pi m})\sin\left(\frac{\pi mx}{2}\right)e^{-m^2\pi^2 t/4}) + 1)$. The plots will be the following:

```python
def f2(x, t):
    n = 10
    s = [(-2/(np.pi*k))*np.sin(np.pi*k*x/2)*np.exp(-k**2*np.pi**2*t/4)for k in range(1, n+1)]
    return sum(s) + 1 -x/2
```
