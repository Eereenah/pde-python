## Problem 1

#### Simulate 3 possible scenarios for the Amazon stock price change for the next 100 days starting from today. Use Geometric Brownian Motion Model for the price:
$\frac{dS_t}{S_t} = \mu \cdot dt + \sigma \cdot dW_t$,  $t = 0, dt, ..., 99\cdot dt$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
mu = 0.06
sigma = 0.14
dt = 1/365
```

```python
s0 = 1765.70 #Amazon stock price as of 28 March 2019, 12:14 
```

Generate $dWt \sim \mathcal{N}(0, dt)$

```python
np.random.seed(0)
dwt1 = np.random.normal(0, 1/365, 100)
np.random.seed(1)
dwt2 = np.random.normal(0, 1/365, 100)
np.random.seed(2)
dwt3 = np.random.normal(0, 1/365, 100)
```

Next, use the formula:

$S_{t+dt} - S_t = S_t(\mu dt + \sigma \mathcal{N(0, dt)})$

```python
def S(mu, sigma, dt, dwt, s0, length):
    s = np.zeros(length)
    s[0] = s0
    for i in range(1, length):
        s[i] = s[i-1] + s[i-1]*(mu*dt + sigma*dwt[i])
    return s
```

```python
x = list(range(100))
```

```python
S1 = S(mu, sigma, dt, dwt1,  s0, 100)
S2 = S(mu, sigma, dt, dwt2,  s0, 100)
S3 = S(mu, sigma, dt, dwt3,  s0, 100)
```

```python
plt.plot(x, S1, x, S2, x, S3)
plt.show()
```
