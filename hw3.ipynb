### c) Define function  $$ f(x) = \begin{cases}
\frac{e^x}{1+e^x}, if \ x < 1.1\\[6pt]
\frac{1}{ln(x)}, if \ x \geq 1.1
\end{cases} $$

### Plot the graph of $f$ in $[-2, 2]$

```python
def f(x):
    if x < 1.1:
        return math.e**x/(1 + math.e**x)
    else:
        return 1/np.log(x)
```

```python
x = np.linspace(-2, 2)
vf = np.vectorize(f)
y = vf(x)
plt.plot(x,y,'-')
plt.show()
```

### d) Plot the graph of $f(x) = x^2, x \in [-\pi, \pi]$ and of its Fourier Series' 10th partial sum on the same graph

```python
def fs(x, n):
    result = 0
    for i in range(1, n+1):
        result = result + 4*((-1)**i)/(i**2)*(math.cos(i*x))
    result = (math.pi**2)/3 + result
    return result
```

```python
fs_x = np.linspace(-math.pi, math.pi)
vfs = np.vectorize(fs)
fs_y = vfs(fs_x, 10)

a = fs_x**2

plt.plot(fs_x,a,'-', color ='b') #the original function
plt.plot(fs_x,fs_y,'-', color ='r') #the approximation
plt.show()
```
